# **LEVEL 1** 
### Task 1: Top Cuisines
1. Determine the top three most common cuisines in the dataset.
2. Calculate the percentage of restaurants that serve each of the top cuisines.

```python
cuisines = df['Cuisines'].dropna().str.split(',').explode().str.strip()
cuisine_counts = cuisines.value_counts()
total = len(df)

print("Top 3 Cuisines:")
for cuisine, count in cuisine_counts.head(3).items():
    pct = (count / total) * 100
    print(f"  {cuisine}: {count} restaurants ({pct:.2f}%)")
```

```python
# City with highest number of restaurants
city_counts = df['City'].value_counts()
print(f"Most restaurants: {city_counts.idxmax()} ({city_counts.max()})")
```

# **LEVEL 1**
### Task 3 : Price Range Distribution
1. Create a histogram or bar chart to
visualize the distribution of price ranges
among the restaurants.

2. Calculate the percentage of restaurants
in each price range category.


```python
labels = {1: 'Budget (1)', 2: 'Mid-range (2)', 3: 'Premium (3)', 4: 'Luxury (4)'}
price_counts = df['Price range'].value_counts().sort_index()

print("Price Range Distribution:")
for pr, count in price_counts.items():
    pct = count / len(df) * 100
    print(f"  {labels[pr]}: {count} ({pct:.1f}%)")

# Bar chart
colors = ['#5DCAA5','#534AB7','#D85A30','#D4537E']
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

axes[0].bar([labels[i] for i in price_counts.index],
            price_counts.values, color=colors)
axes[0].set_title('Restaurants by Price Range')
axes[0].set_ylabel('Count')
axes[0].tick_params(axis='x', rotation=20)

axes[1].pie(price_counts.values,
            labels=[labels[i] for i in price_counts.index],
            colors=colors, autopct='%1.1f%%', startangle=90)
axes[1].set_title('Price Range Share')

plt.tight_layout()
plt.show()
```

### Task 3 : Geographic Analysis

```python
geo = df.dropna(subset=['Latitude','Longitude'])
geo = geo[(geo['Latitude'] != 0) & (geo['Longitude'] != 0)]
print(f"Valid coordinates: {len(geo)} restaurants")

colors = {1:'#5DCAA5', 2:'#534AB7', 3:'#D85A30', 4:'#D4537E'}
labels = {1:'Budget', 2:'Mid-range', 3:'Premium', 4:'Luxury'}

fig, ax = plt.subplots(figsize=(15, 8))
for pr in [1, 2, 3, 4]:
    sub = geo[geo['Price range'] == pr]
    ax.scatter(sub['Longitude'], sub['Latitude'],
               c=colors[pr], label=labels[pr],
               alpha=0.5, s=10)

ax.set_title('Global Restaurant Distribution by Price Range')
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
ax.legend(markerscale=2)
plt.tight_layout()
plt.show()
```

```python
# Cluster analysis with KMeans
from sklearn.cluster import KMeans
import numpy as np

coords = geo[['Latitude','Longitude']].values
kmeans = KMeans(n_clusters=5, random_state=42, n_init=10)
geo = geo.copy()
geo['Cluster'] = kmeans.fit_predict(coords)

fig2, ax2 = plt.subplots(figsize=(15, 8))
for cl in range(5):
    sub = geo[geo['Cluster'] == cl]
    ax2.scatter(sub['Longitude'], sub['Latitude'], alpha=0.5, s=10, label=f'Cluster {cl+1}')
ax2.set_title('Restaurant Clusters (KMeans, k=5)')
ax2.legend()
plt.tight_layout()
plt.show()
```

# **LEVEL 2**
### Task 4 : Restaurants Chains

```python
# Chains = restaurant names appearing more than once
name_counts = df['Restaurant Name'].value_counts()
chains = name_counts[name_counts > 1]
print(f"Total chains found: {len(chains)}")
print(f"Total chain locations: {chains.sum()}")
print("\nTop 15 chains by locations:")
print(chains.head(10))
```

```python
# Ratings & popularity of top chains
chain_df = df[df['Restaurant Name'].isin(chains.index)]
chain_stats = chain_df.groupby('Restaurant Name').agg(
    locations=('Restaurant ID','count'),
    avg_rating=('Aggregate rating','mean'),
    total_votes=('Votes','sum')
).sort_values('locations', ascending=False)

print("\nTop 10 chains — ratings & popularity:")
print(chain_stats.head(10).round(2))
```

```python
# Plot
top10 = chain_stats.head(10)
fig, axes = plt.subplots(1, 2, figsize=(14, 6))
top10['avg_rating'].sort_values().plot(kind='barh', ax=axes[0])
axes[0].set_title('Top Chains — Avg Rating')
top10['total_votes'].sort_values().plot(kind='barh', ax=axes[1])
axes[1].set_title('Top Chains — Total Votes')
plt.tight_layout()
plt.show()
```

# **LEVEL 3**
### Task 1 : Restaurant Reviews

```python
from collections import Counter
import re, matplotlib.pyplot as plt
```

```python
# Using 'Rating text' as a proxy for keyword analysis.

# Positive keywords from high-rated restaurants (>= 4)
pos = df[df['Aggregate rating'] >= 4]['Rating text'].dropna()
neg = df[df['Aggregate rating'] < 3]['Rating text'].dropna()

stopwords = {'the','a','and','is','in','it','of','to','was','for','with','this'}

def top_words(series, n=10):
    words = []
    for text in series:
        words += [w.lower() for w in re.findall(r"\b[a-z]+\b", text)
                  if w.lower() not in stopwords and len(w) > 2]
    return Counter(words).most_common(n)

print("Top positive keywords:")
print(top_words(pos))
print("\nTop negative keywords:")
print(top_words(neg))
```

# **LEVEL 2**
### Task 2 : Votes Analysis

```python
# Highest & lowest votes
print("Top 10 most voted restaurants:")
print(df.nlargest(10, 'Votes')[['Restaurant Name','City','Votes','Aggregate rating']])

print("\nTop 10 least voted restaurants (non-zero):")
low = df[df['Votes'] > 0].nsmallest(10, 'Votes')
print(low[['Restaurant Name','City','Votes','Aggregate rating']])
```

```python
# Correlation
rated = df[df['Aggregate rating'] > 0]
corr = rated[['Votes','Aggregate rating']].corr().iloc[0,1]
print(f"\nPearson correlation (Votes vs Rating): {corr:.4f}")
```

```python
# Scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(rated['Votes'], rated['Aggregate rating'],
            alpha=0.3, s=15)
m, b = np.polyfit(rated['Votes'], rated['Aggregate rating'], 1)
x = np.linspace(0, rated['Votes'].max(), 100)
plt.plot(x, m*x + b, color='#D85A30', linewidth=2, label=f'r = {corr:.2f}')
plt.title('Votes v/s Aggregate Rating')
plt.xlabel('Number of Votes')
plt.ylabel('Aggregate Rating')
plt.legend()
plt.tight_layout()
plt.show()
```

# **LEVEL 3**
### Task 3 : Price Range V/s Delivery & Booking

```python
import warnings
warnings.simplefilter('ignore')

labels = {1:'Budget',2:'Mid-range',3:'Premium',4:'Luxury'}
df['Price Label'] = df['Price range'].map(labels)

# % with delivery & booking per price range
result = df.groupby('Price range').apply(lambda g: pd.Series({
    'Total': len(g),
    'Delivery %': (g['Has Online delivery']=='Yes').mean()*100,
    'Booking %':  (g['Has Table booking']=='Yes').mean()*100,
})).round(1)
result.index = [labels[i] for i in result.index]
print("Service availability by price range:")
print(result)
```

```python
# Plot
x = np.arange(4)
w = 0.35
price_lbls = ['Budget','Mid-range','Premium','Luxury']

fig, ax = plt.subplots(figsize=(11, 6))
ax.bar(x - w/2, result.loc[price_lbls,'Delivery %'], w,
       label='Online Delivery %', color='#534AB7')
ax.bar(x + w/2, result.loc[price_lbls,'Booking %'], w,
       label='Table Booking %', color='#1D9E75')
ax.set_xticks(x)
ax.set_xticklabels(price_lbls)
ax.set_ylabel('Percentage of Restaurants (%)')
ax.set_title('Online Delivery & Table Booking by Price Range')
ax.legend()
plt.tight_layout()
plt.show()
```
